```python
import nbformat
from nbclient import NotebookClient

notebook_file = "google.ipynb"

with open(notebook_file, encoding="utf-8") as f:
    nb = nbformat.read(f, as_version=4)

client = NotebookClient(nb)
client.execute()
```
